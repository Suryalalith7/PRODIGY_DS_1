```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/content/DATA FILE.csv')

# Define the age groups
age_groups = ['20 to 30', '30 to 40', '40 to 50', '50 to 60']

# Define the designation categories
designations = df['Designation'].unique()

# Initialize a dictionary to store the counts for each age group and designation
data_dict = {age_group: {designation: 0 for designation in designations} for age_group in age_groups}

# Iterate through the data and count employees in each age group and designation
for index, row in df.iterrows():
    age = row['Age']
    designation = row['Designation']

    if 20 <= age < 30:
        age_group = '20 to 30'
    elif 30 <= age < 40:
        age_group = '30 to 40'
    elif 40 <= age < 50:
        age_group = '40 to 50'
    else:
        age_group = '50 to 60'

    data_dict[age_group][designation] += 1

# Prepare the data for plotting
age_group_labels = list(data_dict.keys())
designation_labels = list(designations)
data = [[data_dict[age_group][designation] for designation in designation_labels] for age_group in age_group_labels]

# Create a bar chart
fig, ax = plt.subplots()
width = 0.2
x = range(len(age_group_labels))

for i, age_group in enumerate(age_group_labels):
    ax.bar([p + width * i for p in x], data[i], width=width, label=age_group)

ax.set_xlabel('Designation')
ax.set_ylabel('Number of Employees')
ax.set_title('Age Group Distribution by Designation')
ax.set_xticks([p + 1.5 * width for p in x])
ax.set_xticklabels(designation_labels)
ax.legend(age_group_labels)

plt.tight_layout()
plt.show()
```
